```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python

# Paramètres du modèle
r = 0.05
K = 100_000
P0 = 1_000
h = 1  # pas mensuel

# Initialisation
t = 0
P = P0
temps = [t]
clients = [P]

# Simulation avec Euler
while P < 50000:
    dP = r * P * (1 - P/K)
    P += dP * h
    t += h
    temps.append(t)
    clients.append(P)

# Résultat
print(f"Temps pour atteindre 50 000 clients : {t} mois")

# Visualisation
plt.figure(figsize=(10,6))
plt.plot(temps, clients)
plt.xlabel('Mois')
plt.ylabel('Nombre de clients')
plt.title('Croissance des clients (modèle logistique)')
plt.grid(True)
plt.show()
```
